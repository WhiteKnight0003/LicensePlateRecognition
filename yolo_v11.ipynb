```python
from ultralytics import YOLO
import shutil , os
```

```python
# Load pretrained YOLOv8 nano model
model = YOLO("yolo11m.pt")

currently_path = os.getcwd()
runs_path = currently_path + "/runs"
results_path = runs_path + "/detect"
if os.path.exists(runs_path):
   shutil.rmtree(runs_path)
os.makedirs(runs_path)
os.makedirs(results_path)

# Train
results = model.train(
   data=dataset.location + "/data.yaml",
   # use dataset path from Roboflow
   epochs=50,
   imgsz=640,
   batch=4,
   workers=3,
   device=0,
   project = results_path,  # ƒê·∫∑t ƒë∆∞·ªùng d·∫´n tuy·ªát ƒë·ªëi ƒë·∫øn th∆∞ m·ª•c b·∫°n mu·ªën
   name = 'train'  # T√™n th∆∞ m·ª•c con cho l·∫ßn ch·∫°y n√†y
)

# Copy weights to a safe folder
os.makedirs("saved_models", exist_ok=True)
shutil.copy("runs/detect/train/weights/best.pt", "saved_models/license_plate_best.pt")
shutil.copy("runs/detect/train/weights/last.pt", "saved_models/license_plate_last.pt")
print("Weights saved in saved_models/")
```
